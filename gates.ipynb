```python
from qiskit import *
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_histogram, array_to_latex, plot_bloch_multivector
from qiskit_aer import AerSimulator, StatevectorSimulator
```

```python
# Flipping bits (Pauli's X Gate)

# Here we know that this is a qbit initialized at state 0
# Classical register is needed to store the measured position of the qbit when it collapses
qr = QuantumRegister(1)
cr = ClassicalRegister(1)
circuit = QuantumCircuit(qr, cr)

# X gate, acts like a classical NOT gate
circuit.x(qr)

circuit.measure(qr, cr)

simulator = StatevectorSimulator()
compiledCircuit = simulator.run(circuit).result()
compiledCircuit.data()
statevec = compiledCircuit.get_statevector()
array_to_latex(statevec)
plot_bloch_multivector(statevec)
```

```python
# Hadamard's Gate

quantumRegister = QuantumRegister(1)
classicalRegister = ClassicalRegister(1)
circuit = QuantumCircuit(quantumRegister, classicalRegister)

#Hadamard gate
circuit.h(quantumRegister)

sim = StatevectorSimulator()
res = sim.run(circuit).result()
res.data()
stateVec = res.get_statevector()
array_to_latex(stateVec)
plot_bloch_multivector(stateVec)
```

```python
quantumRegister = QuantumRegister(1)
classicalRegister = ClassicalRegister(1)
circuit = QuantumCircuit(quantumRegister, classicalRegister)

#Hadamard gate
circuit.h(quantumRegister)
circuit.measure(quantumRegister, classicalRegister)

simulator = AerSimulator()
compiledCircuit = transpile(circuit, simulator)
result = simulator.run(compiledCircuit, shots = 1200).result()
counts = result.get_counts()
plot_histogram(counts)
```

```python
# CX gate

qr = QuantumRegister(2)
cr = ClassicalRegister(2)
circuit = QuantumCircuit(qr, cr)

# CX gate applies X gate to the socond qbit if the control qbit is 1
circuit.x(qr) #Pauli's X gate; flips bits
circuit.h(qr[0]) #Hadamard gate
circuit.cx(qr[0], qr[1])
circuit.measure(qr, cr)

simulator = AerSimulator()
compiledCircuit = transpile(circuit, simulator)
result = simulator.run(compiledCircuit, shots = 1200).result()
counts = result.get_counts()
print(counts)
plot_histogram(counts)
```

```python
#CX with Hadamard

quantumRegister = QuantumRegister(2)
classicalRegister = ClassicalRegister(2)
circuit = QuantumCircuit(quantumRegister, classicalRegister)

#Hadamard gate and CNOT gate entangle qbits
circuit.h(quantumRegister[0])
circuit.cx(quantumRegister[0], quantumRegister[1])
circuit.measure(quantumRegister, classicalRegister)

simulator = AerSimulator()
compiledCircuit = transpile(circuit, simulator)
result = simulator.run(compiledCircuit, shots = 2000).result()
counts = result.get_counts()
print(counts)
#we get all 0's because we haven't applied any quantum operations to the qbit
```

```python
# Flipping bits with a phase flip too (Pauli's Y Gate)

# Here we know that this is a qbit initialized at state 0
# Classical register is needed to store the measured position of the qbit when it collapses
qr = QuantumRegister(1)
cr = ClassicalRegister(1)
circuit = QuantumCircuit(qr, cr)

# Y gate, flips the qbit and adds a phase of qbit
circuit.y(qr)

circuit.measure(qr, cr)

simulator = StatevectorSimulator()
compiledCircuit = simulator.run(circuit).result()
compiledCircuit.data()
statevec = compiledCircuit.get_statevector()
array_to_latex(statevec)
plot_bloch_multivector(statevec)
```

```python
# Flipping phase (Pauli's Z Gate)

# Here we know that this is a qbit initialized at state 0
# Classical register is needed to store the measured position of the qbit when it collapses
qr = QuantumRegister(1)
cr = ClassicalRegister(1)
circuit = QuantumCircuit(qr, cr)

# Z gate, flips the phase
circuit.y(qr)

circuit.measure(qr, cr)

simulator = StatevectorSimulator()
compiledCircuit = simulator.run(circuit).result()
compiledCircuit.data()
statevec = compiledCircuit.get_statevector()
array_to_latex(statevec)
plot_bloch_multivector(statevec)
```
